```python
import csv
csv.field_size_limit(1000000000)
from gensim import corpora, models, similarities
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)
```

```python
#load the data
f = open('/Users/mayli/Desktop/state-of-the-union.csv')
csv_f = csv.reader(f)

#as a list of strings, one per document
documents = []
#Create Decade Dictionary for each decade with docids
year0 = 1790
docid_list = []
dict_decade = {}
docid = 0

for row in csv_f:
    documents.append(str(row[1]))
    if int(row[0]) - int(year0)<10:
        #print(row[0])
        docid_list.append(docid)
        dict_decade[year0] = docid_list
        #print(dict_decade)
        
    else:
        year0 = year0 +10
        docid_list = []
        docid_list.append(docid)
        dict_decade[year0] = docid_list
    #print(dict_decade)
    docid = docid + 1
    
for key,value in dict_decade.items():
    print('{key}:{value}'.format(key = key, value = value))
print(len(documents))
```

```python
#load stopwords 
f1 = open('/Users/mayli/Desktop/stopwords-es.csv')
csv_f1 = csv.reader(f1)

stopwords = []
for row in csv_f1:
    stopwords.append(row[0])

stoplist = set(stopwords)
```

```python
# remove common words and tokenize
texts = [[word for word in document.lower().split() if word not in stoplist]
          for document in documents]
```

```python
# remove words that appear only once
from collections import defaultdict
frequency = defaultdict(int)
for text in texts:
    for token in text:
        frequency[token] += 1

texts = [[token for token in text if frequency[token] > 1] for text in texts]
```

```python
# build the dictionary, a dict of (question, answer) pair and save it
dictionary = corpora.Dictionary(texts)
```

```python
# transform the whole texts to vector(sparse) and save it
corpus = [dictionary.doc2bow(text) for text in texts]
```

```python
#transform to Tf-Idf corpus
tfidf = models.TfidfModel(corpus) 
corpus_tfidf = tfidf[corpus]

dict_doc = {}
docid = 0

for doc in corpus_tfidf:
    dict_doc[docid] = doc
    docid = docid + 1
    #print(doc)
    #print('\n')
    
#for key,value in dict_doc.items():
    #print('{key}:{value}'.format(key = key, value = value))
    #print('\n')
```
